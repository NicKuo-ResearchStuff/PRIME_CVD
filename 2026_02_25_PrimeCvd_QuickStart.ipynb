```python
import pandas as pd
import zipfile
import requests
import io

url = "https://ndownloader.figshare.com/files/62130498"

r = requests.get(url)
r.raise_for_status()

zip_bytes = io.BytesIO(r.content)

with zipfile.ZipFile(zip_bytes) as z:
    file_list = z.namelist()
    print("Files inside zip:", file_list)

    dataframes = {}
    for file_name in file_list:
        if file_name.endswith(".csv"):
            with z.open(file_name) as f:
                df = pd.read_csv(f)

                df = df.drop(columns=["Unnamed: 0"], errors="ignore")

                dataframes[file_name] = df

patient_master = next(df for name, df in dataframes.items() if "MasterSummary" in name)
chronic_diseases = next(df for name, df in dataframes.items() if "ChronicDiseases" in name)
meas_and_path = next(df for name, df in dataframes.items() if "MeasAndPath" in name)
```
